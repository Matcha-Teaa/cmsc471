```python
import pandas as pd
```

```python
raw = pd.read_csv("./data.csv")
raw = raw.replace(',','', regex=True)
raw = raw.replace('%','', regex=True)
raw = raw.replace('-', '')

raw = raw[4:178].reset_index()

for col in raw.columns[3:]:
    raw[col] = pd.to_numeric(raw[col], errors='coerce')

df = pd.DataFrame()

df["country"] = raw["ISO Country code"]

df["GDP \n($ USD billions PPP) (2018)"] = raw.iloc[:,15]
df["GDP \n($ USD billions PPP) (2019)"] = raw.iloc[:,16]
df["GDP \n($ USD billions PPP) (2020)"] = raw.iloc[:,17]
df["GDP \n($ USD billions PPP) (2021)"] = raw.iloc[:,18]

df["GDP per capita in $ (PPP) (2018)"] = raw.iloc[:,19]
df["GDP per capita in $ (PPP) (2019)"] = raw.iloc[:,20]
df["GDP per capita in $ (PPP) (2020)"] = raw.iloc[:,21]
df["GDP per capita in $ (PPP) (2021)"] = raw.iloc[:,22]

df["health expenditure \n% of GDP (2014)"] = raw.iloc[:,26]
df["health expenditure \n% of GDP (2015)"] = raw.iloc[:,27]
df["health expenditure \n% of GDP (2016)"] = raw.iloc[:,28]
df["health expenditure \n% of GDP (2017)"] = raw.iloc[:,29]
df["health expenditure \n% of GDP (2018)"] = raw.iloc[:,30]
df["health expenditure \n% of GDP (2019)"] = raw.iloc[:,31]
df["health expenditure \n% of GDP (2020)"] = raw.iloc[:,32]
df["health expenditure \n% of GDP (2021)"] = raw.iloc[:,33]

df["health expenditure \nper person (2015)"] = raw.iloc[:,35]
df["health expenditure \nper person (2018)"] = raw.iloc[:,36]
df["health expenditure \nper person (2019)"] = raw.iloc[:,37]

df["unemployment (%) (2018)"] = raw.iloc[:,49]
df["unemployment (%) (2021)"] = raw.iloc[:,50]

df["Military Spending as % of GDP (2021)"] = raw.iloc[:,85]
df["Military Spending as % of GDP (2019)"] = raw.iloc[:,86]
```

```python
df.to_csv("./data-processed.csv", index=False)
```
